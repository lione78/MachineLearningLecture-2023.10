### 다중 선형회귀 - 보스톤 주택 가격

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

- 데이터 탐색

```python
df = pd.read_csv('data/Boston_house.csv')
df.head()
```

- Feature별 산점도와 선형회귀선

```python
plt.figure(figsize=(4, 3))
sns.regplot(data=df, x='AGE', y='Target');
```

```python
fig, axs = plt.subplots(3, 4, figsize=(16, 14), sharey=True)
for i, feature in enumerate(df.columns[1:-1]):
    row, col = i // 4, i % 4
    sns.regplot(data=df, x=feature, y='Target', ax=axs[row, col])
```

- Feature별 결정 계수와 산점도

```python
from sklearn.linear_model import LinearRegression
fig, axs = plt.subplots(3, 4, figsize=(16, 14), sharey=True)
for i, feature in enumerate(df.columns[1:-1]):
    row, col = i // 4, i % 4
    ax = axs[row, col]
    sns.regplot(data=df, x=feature, y='Target', ax=ax)
    lr = LinearRegression()
    X = df[feature].values.reshape(-1, 1)       # train_test_split전에는 reshape(-1, 1) 해 줌. W[[,]] * X[[],[]] + b
    lr.fit(X, df.Target.values)
    score = lr.score(X, df.Target.values)
    ax.set_title(f'{feature}: {score:.4f}')
```

- 다중 선형회귀

```python
# Categorical subset data에서 사용. 연속적인 데이터에서 train_test_split하기 어렵지만
# 전략을 가지고 하면 할 수 있음.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    df.iloc[:, :-1].values, df.iloc[:, -1].values, test_size=0.2, random_state=2023     # 2차원이므로 reshape한면 암됨.
)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# 결정계수(R Squared)
lr.score(X_test, y_test)
```

```python
# 가중치, 바이어스
lr.coef_, lr.intercept_
```

- 회귀식
    
    Target = -0.0053 * AGE + -0.0071 * B + 3.63 * RM + ...... + 2.42 * CHAS + 35.86

- 개별 테스트 데이터에 적용

```python
X_test[0], y_test[0]
```

```python
# weight, bias를 이용하여 직접 계산
np.dot(lr.coef_, X_test[0]) + lr.intercept_
```

```python
# predict() method 사용
y_hat = lr.predict(X_test[0].reshape(1, -1))
y_hat
```

```python
# 10개 테스트 테이터에 적용
y_hat = lr.predict(X_test)
for i in range(10):
    print(f'실제값: {y_test[i]}, \t예측값: {y_hat[i]:.2f}')
```

- 회귀는 선형회귀 외에도 다른 모델이 많음

- SVR

```python
from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_squared_error
```

```python
svr = SVR()
svr.fit(X_train, y_train)
y_hat_svr = svr.predict(X_test)
r2_sv = r2_score(y_test, y_hat_svr)
mse_sv = mean_squared_error(y_test, y_hat_svr)
r2_sv
```

- Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(random_state=2023)
rfr.fit(X_train, y_train)
y_hat_rfr = rfr.predict(X_test)
r2_rf = r2_score(y_test, y_hat_rfr)
mse_rf = mean_squared_error(y_test, y_hat_rfr)
r2_rf
```

- XGBoost

```python
from xgboost import XGBRegressor
xgbr = XGBRegressor()
xgbr.fit(X_train, y_train)
y_hat_xgb = xgbr.predict(X_test)
r2_xgb = r2_score(y_test, y_hat_xgb)
mse_xgb = mean_squared_error(y_test, y_hat_xgb)
r2_xgb
```

- Linear Regression

```python
r2_lr = r2_score(y_test, y_hat)
mse_lr = mean_squared_error(y_test, y_hat)
```

- 비교

```python
print('\t\tLR\tSVM\tRF\tXGBoost')
print(f'R Squared\t{r2_lr:.4f}\t{r2_sv:.4f}\t{r2_rf:.4f}\t{r2_xgb:.4f}')
print(f'MSE\t\t{mse_lr:.2f}\t{mse_sv:.2f}\t{mse_rf:.2f}\t{mse_xgb:.2f}')        # MSE는 소숫점 아래가 중요치 않다.
```

```python
# rf : result frame
rf = pd.DataFrame({
    'y': y_test, 'LR': y_hat, 'SVM': y_hat_svr, 'RF': y_hat_rfr, 'XGBoost': y_hat_xgb 
})
rf.head(10)
```
